```python
imdb_df = pd.read_csv("./zippedData/imdb.title.basics.csv.gz", compression="infer")
imdb_df.head()
```

```python
# this will get them into a list
genres_list = imdb_df['genres'].str.lower().str.split(",").to_list()
```

```python
# how can we get these out in just 1 single list
g_list = []
for genre_list in genres_list:
    try:
        g_list.extend(genre_list)
    except:
        continue
```

# let's dummify our genres


```python
# this is converting all of the lists into a Series so pandas can hash through it
# so you iterate or index through a list
# you has through a dictionary

# we're converting our column of lists to something that is hashable
imdb_genres_df = pd.get_dummies(imdb_df['genres'].str.split(",").apply(pd.Series).stack(),
               prefix='genre',
               prefix_sep="_").sum(level=0)
# this does not include nulls

imdb_genres_df.head()
```

```python
imdb_genres_df.plot(kind='bar')
plt.show()
```
