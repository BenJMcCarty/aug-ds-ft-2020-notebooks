# Typically APIs have endpoints
* a place to submit requests (front door)
    * typically looks like a URL
    * make a request to this URL and include parameters
        * parameters come after `/?param1=param1&`
* endpoint takes in the request (and sometimes credentials)
    * sends a response (typically a json response)

# actual example using twitter api

```python
import tweepy as tw
```

```python
with open("/Users/rafael/.ssh/twitter_app00.json") as f:
    d = json.load(f)
    auth = tw.OAuthHandler(consumer_key=d['consumer_key'], 
                           consumer_secret=d['consumer_secret'])
    auth.set_access_token(key=d['access_token'], secret=d['access_token_secret'])
    del d
```

```python
api = tw.API(auth_handler=auth)
```

```python
api.update_status("Currently teaching about APIs. Sending this from Python! Hello World!!!!")
```

```python
import os


class RickAndMortyAPI():
    
    
    def __init__(self, base_url="https://rickandmortyapi.com/api/"):
        self.base_url = base_url
    
    
    def get_alive_characters(self):
        endpoint = os.path.join(self.base_url, "character")
        params = {"status":"alive"}
        res = requests.get(endpoint, params=params)
        return res.json()
    
    
    def query_characters(self, **params):
        """
        Available parameters:

        name: filter by the given name.
        status: filter by the given status (alive, dead or unknown).
        species: filter by the given species.
        type: filter by the given type.
        gender: filter by the given gender (female, male, genderless or unknown).
        """
        endpoint = os.path.join(self.base_url, "character")
        res = requests.get(endpoint, params=params)
        return res.json()
```

```python
rm_api = RickAndMortyAPI()
```

```python
rm_api.query_characters(params={"status":"alive"})
```
