```python
for mu, color in zip([50, 60], ['red', 'blue']):
    dist = scs.norm(loc=mu, scale=10)
    x_axis = np.linspace(dist.ppf(0.01), dist.ppf(.99), 1000)
    probs = dist.pdf(x_axis)
    plt.plot(x_axis, probs, c=color, label=mu)
    plt.legend()
    plt.title(f"Norm Dist with Mu={mu}")
plt.show()
```

changing the mean of a normal distribution shifts the distribution

```python
for std, color in zip([10, 20], ['red', 'blue']):
    dist = scs.norm(loc=50, scale=std)
    x_axis = np.linspace(dist.ppf(0.01), dist.ppf(.99), 1000)
    probs = dist.pdf(x_axis)
    plt.plot(x_axis, probs, c=color, label=std)
    plt.legend()
    plt.title(f"Norm Dist with std={std}")
plt.show()
```

standard deviation changing, changes the width. standard deviation measures the spread of the data. standard deviation measures the centralization of the data 

```python
dist_norm = np.random.normal(50, 10, 1000)
plt.hist(dist_norm, bins=20)
plt.show()
```

this data is not quite symmetrical. there's a long left tail. negative skew.

blob on left side around 35/70.

all hail the grand eye (triangular)

```python
# skewness and kurtosis
# assess normality


# these measurements are all in terms of the standard normal std
# std = 1.0, mu=0 this is the standard normal distribution
# skewness measures the left/right to the mean of a distribution
# kurtosis measures inside/outside to the mean of a distribution

scs.skew(dist_norm), scs.kurtosis(dist_norm, fisher=False)

# ideal values are skew=0 and kurt=3 for kurtosis without fisher
```

```python
scs.skew(dist_norm), scs.kurtosis(dist_norm, fisher=True)

# ideal values are skew=0 and kurt=0 for kurtosis with fisher
```

# Zscores 

```python
height = np.random.normal(80, 4, 1000)
shoe_sizes = np.random.normal(15, 0.8, 1000)

plt.hist(height, label="height", color='r')
plt.hist(shoe_sizes, label="shoe size", color='b')
plt.legend()
plt.show()
```

# zscores formula

$ z = \frac{x - \mu}{\sigma} $

in terms of standard deviation

how many standard deviations from the mean is x?

notice this is using the mean and std, which means it's assuming X is normally distributed

```python
z_score_76 = (73-80)/4
```

```python
def zscore(dist):
    return (dist-dist.mean())/dist.std()
```

```python
z_height = zscore(height)
z_shoesize = zscore(shoe_sizes)
```

```python
# notice the mean has moved to 0
# notice the standard deviation is 1
plt.hist(z_height)
plt.show()
```

```python
# this is unitless
plt.hist(z_height, label="height", color='r', alpha=0.5)
plt.hist(z_shoesize, label="shoe size", color='b', alpha=0.5)
plt.legend()
plt.show()
```

# how to test for normality


```python
# h0: X is normal
# Ha: X is not normal
pvalue, tstat = scs.shapiro(height)
pvalue
# pvalue of 0.99
# pvalue > 0.05 which we means we can fail to reject the null

# pvalue >0.05 means data is normal
```
